```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def julia_set(width, height, xmin, xmax, ymin, ymax, c, max_iter):
    x, y = np.linspace(xmin, xmax, width), np.linspace(ymin, ymax, height)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    img = np.zeros(Z.shape, dtype=float)

    for i in range(max_iter):
        Z = Z**2 + c
        mask = np.abs(Z) < 1000
        img += mask

    img[img == max_iter] = 0
    img = np.log(img + 1)

    return img

# Set parameters
width, height = 800, 800
xmin, xmax = -2, 2
ymin, ymax = -2, 2
c = complex(-0.7, 0.27015)
max_iter = 100

# Create and display the Julia set
julia_img = julia_set(width, height, xmin, xmax, ymin, ymax, c, max_iter)
plt.imshow(julia_img, cmap='twilight')
plt.axis('off')
plt.show()
```
