```python
import matplotlib.pyplot as plt
import numpy as np

def sierpinski_triangle(order, size):
    points = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])
    result = np.array([[0.5, np.sqrt(3) / 4]])

    for _ in range(size):
        chosen = np.random.choice(points.shape[0], 1, replace=True)
        new_point = (result[-1] + points[chosen][0]) / 2
        result = np.concatenate([result, [new_point]])

    return result

# Set parameters
order = 5
size = 50000

# Create and display the Sierpinski Triangle
triangle_points = sierpinski_triangle(order, size)
plt.scatter(triangle_points[:, 0], triangle_points[:, 1], s=1, c='black')
plt.axis('equal')
plt.axis('off')
plt.show()
```
