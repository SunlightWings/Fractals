```python
import matplotlib.pyplot as plt
import numpy as np

def draw_succulent(ax, x, y, size, angle, depth):
    if depth == 0:
        return
    else:
        for i in range(6):
            x_end = x + size * np.cos(np.radians(angle + i * 60))
            y_end = y + size * np.sin(np.radians(angle + i * 60))

            ax.plot([x, x_end], [y, y_end], color='green', lw=1)

            draw_succulent(ax, x_end, y_end, size * 0.7, angle + 30, depth - 1)

def draw_succulent_fractal(order):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect('equal', adjustable='datalim')
    ax.axis('off')

    start_x, start_y = 0, 0
    start_size = 1
    start_angle = 90  # Initial angle, you can adjust as needed

    draw_succulent(ax, start_x, start_y, start_size, start_angle, order)

    plt.show()

# Set the order to control the complexity of the fractal
order = 5  # You can adjust this value to control the complexity

draw_succulent_fractal(order)
```
